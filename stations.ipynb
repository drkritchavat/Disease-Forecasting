```python
##HTTP GET Requests to TMD API for station information
import requests
import pandas as pd
sta = requests.get('https://data.tmd.go.th/api/Station/v1/?uid=demo&ukey=demokey&format=json')

##Parsing JSON and store it with dictionary format
sta_dict = sta.json()
sta_dict.keys()
header = sta_dict['header']
header
#### JSON Header #####
```

```python
###Information of the first station###
sta_dict['Station'][0]
```

```python
###Storing with dataframe and save as an csv file
columns = list(sta_dict['Station'][0].keys())[:-3]
columns
df = pd.DataFrame(sta_dict['Station'])[columns]
df.sample(10)
```

```python
df.to_csv('sta.csv')
```
